## Caso 1: Partiendo de que se conocen los enfrentamientos
En este caso, suponemos que los 10 partidos ya están definidos y únicamente debemos decidir en qué horario se disputa cada uno.

Cada uno de los 10 partidos puede asignarse a cualquierea de los 10 horarios disponibles, permitiendo además que varios partidos coincidan en la misma franja.

Esto corresponde a **variaciones con repetición**, por lo tanto el número total de posibilidades es $10^{10}$

Sabemos que existe la **restricción** de que debe haber al menos un partido el viernes a las 20h y al menos un partido el lunes a las 20h.

Entonces, el cálculo final vendría dado por:

- Total: $10^{10}$
- Restar las posibilidades que no incluyen el viernes(20h): $9^{10}$
- Restar las posibilidades que no incluyen el lunes(20h): $9^{10}$
- Sumar las posibilidades que no incluyen ni viernes(20h) ni lunes(20h): $8^{10}$

Posibilidades = $10^{10}$ - $2*9^{10}$ + $8^{10}$ = **4.100.173.022**

## Caso 2: Partiendo de que no se conocen los enfrentamientos
En este escenario, además de asignar horarios, también es necesario generar los emparejamientos de los equipos para formar los 10 partidos de la jornada.

Dado que hay 20 equipos, el problema consiste en dividirlos en 10 parejas, sin repetir estos equipos y sin distinguir el orden local/visitante dentro de cada partido (A vs B es equivalente a B vs A).

Sin aplicar las restricciones de un partido el viernes y otro el lunes, el número de formas de emparejar 20 equipos en 10 partidos, sin repetir, y sin distinguir el orden dentro de cada partido, es:

$N_{\text{partidos}}=\frac{20!}{2^{10}\,10!}$

Una vez definidos los partidos,se deben asignar los horarios cumpliendo las **restricciones** de que haya un partido el viernes a las 20h y un partido el lunes a las 20h.

$N_{\text{horarios}} = 10^{10} - 2\cdot 9^{10} + 8^{10}$

Por lo tanto, el número total de de posibilidades (emparejamientos más asignación de horarios) es:

$N_{\text{total}} = \frac{20!}{2^{10}\,10!}\,\Bigl(10^{10} - 2\cdot 9^{10} + 8^{10}\Bigr)$ = $2.68* 10^{21}$

## ¿Cual es la función objetivo?

## ¿Es un problema de maximización o minimización?

La función objetivo del problema es **maximizar la audiencia total de la jornada**, ya que el enunciado establece que los partidos deben asignarse a horarios de manera que la audiencia sea lo mayor posible.

Para la solución $s$, la audiencia total (en millones de espectadores) se define como:

$f(s) = \sum_{i=1}^{10} A_i \cdot p(h_i) \cdot c(h_i, s)$


donde:
- $A_i$ es la audiencia base (en millones) del partido $i$, determinada por la categoría de los equipos que se enfrentan.
- $h_i$ es el horario asignado al partido $i$.
- $p(h_i)$ es el coeficiente de ponderación del horario, que refleja la reducción de audiencia respecto al mejor horario posible (sabado a las 20h).
- $c(h_i, s)$ es el factor de corrección por coincidencias, que depende del número de partidos que comparten el mismo horario $h_i$ dentro la solución $s$

Por lo tanto, el problema se formula como un **problema de maximización**, donde el objetivo es encontrar la asignación de horarios que maximice el valor de $f(s)$.

## Definimos todos los datos y funciones compartidas

```
# Guardamos los horarios del enunciado en una lista
HORARIOS = ["V20", "S12", "S16", "S18", "S20", "D12", "D16", "D18", "D20", "L20"]

# Guardamos la audiencia base (en millones) en un diccionario
AUDIENCIA_BASE = {
    ("A", "A"): 2.0,
    ("A", "B"): 1.3,
    ("B", "A"): 1.3,
    ("A", "C"): 1.0,
    ("C", "A"): 1.0,
    ("B", "B"): 0.9,
    ("B", "C"): 0.75,
    ("C", "B"): 0.75,
    ("C", "C"): 0.47,
}

# Guardamos en un diccionario que categoría tiene cada equipo
TEAM2CAT = {
    "Real Madrid": "A",
    "Barcelona": "A",
    "R.Sociedad": "A",

    "Celta": "B",
    "Valencia": "B",
    "Athletic": "B",
    "Villarreal": "B",
    "Alavés": "B",
    "Levante": "B",
    "Espanyol": "B",
    "Sevilla": "B",
    "Betis": "B",
    "Atlético": "B",
    "Getafe": "B",

    "Mallorca": "C",
    "Eibar": "C",
    "Leganés": "C",
    "Osasuna": "C",
    "Granada": "C",
    "Valladolid": "C",
}

# Guardamos la ponderación de cada horario en un diccionario
PONDERACION_HORARIO = {
    "V20": 0.4,
    "S12": 0.55,
    "S16": 0.7,
    "S18": 0.8,
    "S20": 1.0,
    "D12": 0.45,
    "D16": 0.75,
    "D18": 0.85,
    "D20": 1.0,
    "L20": 0.4,
}

# Guardamos la penalización por cada partido en la misma franja en un diccionario
PENALIZACION = {
    0: 0.0,
    1: 0.25,
    2: 0.45,
    3: 0.60,
    4: 0.70,
    5: 0.75,
    6: 0.78,
    7: 0.80,
    8: 0.80,
}

# Construimos una lista de tuplas (partidos) para el test de Fuerza Bruta
PARTIDOS_EJEMPLO = [
    ("Celta", "Real Madrid"),
    ("Valencia", "R.Sociedad"),
    ("Mallorca", "Eibar"),
    ("Athletic", "Barcelona"),
    ("Leganés", "Osasuna"),
    ("Villarreal", "Granada"),
    ("Alavés", "Levante"),
    ("Espanyol", "Sevilla"),
    ("Betis", "Valladolid"),
    ("Atlético", "Getafe"),
]
```

```
# DEFINICIÓN DE FUNCIONES AUXILIARES COMPARTIDAS


def categorias_de_partidos(partidos, team2cat):
    """
    Convertimos los partidos en una lista de tuplas de categorías asociadas a los equipos
    """
    cats = []
    for local, visit in partidos:
        if local not in team2cat:
            raise KeyError(f"Falta categoría para el equipo: {local}")
        if visit not in team2cat:
            raise KeyError(f"Falta categoría para el equipo: {visit}")
        cats.append((team2cat[local], team2cat[visit]))
    return cats
```


## Diseña un algoritmo para resolver el problema por fuerza bruta

Para el diseño del algoritmo de fuerza bruta se va a partir de la hipótesis de que se conocen los enfrentamientos de la jornada.

```
# DEFINICIÓN DE FUNCIONES AUXILIARES

def calcular_audiencia_total(categorias, horarios, solucion_horarios):
    """
    Calcula la audiencia total para una asignación concreta de horarios.

    Parámetros:
    - categorias: listas de tuplas (cat_local, cat_visit) por partido.
    - horarios: lista de horarios disponibles (strings).
    - solucion_horarios: lista de horarios por partido, donde solucion_horarios[i] es el horario del partido i.

    Devuelve:
    - Float: audiencia_total
    """
    # Contamos cuántos partidos hay por horario
    conteo = {}
    for h in horarios:
        conteo[h] = 0

    # Sumamos 1 por cada partido asignado a un horario
    for h in solucion_horarios:
        conteo[h] = conteo[h] + 1

    # Recorremos cada partido i y sumamos su contribución:
    # contribuciones_partido_i = audiencia_base(cat_partido_i) * ponderacion(h) * coeficiente_penalizacion(h)
    audiencia_total = 0.0

    for i, (cat_local, cat_visit) in enumerate(categorias):
        h = solucion_horarios[i]

        # Audiencia base por categorías
        base = AUDIENCIA_BASE[(cat_local, cat_visit)]

        # Ponderación por horario
        ponderacion = PONDERACION_HORARIO[h]

        # Determinamos cuántos partidos compiten con este en el mismo horario
        k = conteo[h] - 1

        # Ajustamos k al rango permitido por la tabla penalizacion
        k = max(0, min(k, 8))

        # Calculamos el coeficiente de penalización por coincidencias
        coeficiente_coinc = 1.0 - PENALIZACION[k]

        # Sumamos contribución del partido i a la audiencia total
        audiencia_total += base * ponderacion * coeficiente_coinc

    return audiencia_total


def siguiente_vector_base_b(indices, base):
        """
        Incrementa la lista 'indices' como si fuera un número en base 'base'.

        Su utilidad es enumerar todas las posibles combinaciones de horarios.

        Parámetros:
        - indices: lista de enteros (cada uno en [0, base-1]).
        - base: la cantidad de horarios disponibles (entero).

        Devuelve:
        - Se modifica la lista indices in-place.
        - True: si se pudo incrementar indices (existen más combinaciones).
        - False: si no se pudo incrementar indices (no existen más combinaciones).
        """
        # Empieza por el último digito de la lista
        i = len(indices) - 1
        while i >= 0:
            indices[i] += 1
            # Si no llegamos a 'base', ya quedó bien incrementado el digito en la posicion i
            if indices[i] < base:
                return True
            # Si llegó a 'base', vuelve ese digito a 0 y seguimos con el dígito anterior
            indices[i] = 0
            i -= 1
        # Si nos pasamos del primer dígito no hay mas combinaciones
        return False
```

```
# DEFINICIÓN DE FUERZA BRUTA

def fuerza_bruta(categorias, horarios, exigir_v20_l20=True):
    """
    Recorre TODAS las asignaciones posibles de horarios.

    Parámetros:
    - categorias: listas de tuplas de longitud n
    - horarios: lista de horarios disponibles (strings).
    - exigir_v20_l20: si True aplica la restricción "al menos un V20 y un L20" en la solución final.

    Devuelve:
    - (mejor_solucion, mejor_audiencia)
      siendo mejor_solucion la lista de horarios asignados a cada partido.
    """
    # n = cantidad de partidos
    n = len(categorias)
    # b = cantidad de horarios
    b = len(horarios)

    # Guardamos la mejor solución encontrada hasta ahora
    mejor_solucion = None
    mejor_audiencia = -1.0

    # Arrancamos con todos los partidos en el primer horario (índice 0)
    indices = [0] * n

    # Recorremos todas las combinaciones posibles del vector indices
    hay_mas = True
    while hay_mas:

        # Convertimos los indices en una solucion concreta
        solucion = [horarios[idx] for idx in indices]

        # Comprobamos la restricción: "Debe haber al menos un V20 y un L20"
        if exigir_v20_l20:
            # Si no estan en la solución actual, no calculamos audiencia
            # y avanzamos a la próxima solución
            if ("V20" not in solucion) or ("L20" not in solucion):
                hay_mas = siguiente_vector_base_b(indices, b)
                continue

        # Calculamos la audiencia total
        aud = calcular_audiencia_total(categorias, horarios, solucion)

        # Si encontramos una mejor solución, la guardamos
        if aud > mejor_audiencia:
            mejor_audiencia = aud
            mejor_solucion = solucion[:]

        # Avanzamos a la siguiente combinación
        hay_mas = siguiente_vector_base_b(indices, b)

    return mejor_solucion, mejor_audiencia
```

Como vimos anterioremente, si los enfrentamiento de la jornada ya están definidos y aplicamos la restricción de que debe existir al menos un partido en V20 y al menos uno en L20, el número de combinaciones válidas es:

Posibilidades = $10^{10}$ - $2*9^{10}$ + $8^{10}$ = 4.100.173.022

Esto implica que el algoritmo de fuerza bruta debería evaluar más de cuatro mil millones de combinaciones, lo que lo vuelve computacionalmente inviable en la práctica para una jornada completa.

Por este motivo, para comprobar que la implementación del algoritmo es correcta, utilizamos **un caso de prueba reducido** con 4 partidos y 4 horarios, lo que genera:

Posibilidades del Test = $4^{4}$ = 256

Este tamaño sí permite ejecutar el algoritmo completo en un tiempo razonable y verificar que calcula correctamente la audiencia y encuentra la mejor solución dentro del espacio de búsqueda.

```
def prueba_fuerza_bruta():
    # Armamos un pequeño test para que fuerza bruta sea ejecutable
    # 4 partidos, 4 horarios => 4^4 = 256 combinaciones

    partidos_test = PARTIDOS_EJEMPLO[:4]
    categorias_test = categorias_de_partidos(partidos_test, TEAM2CAT)

    # Tomamos un subconjunto de horarios
    horarios_test = ["V20", "S20", "D20", "L20"]

    # Ejecutamos fuerza bruta SIN restricciones
    solucion_test_sin, audiencia_test_sin = fuerza_bruta(
        categorias_test,
        horarios_test,
        exigir_v20_l20=False
    )

    print("TEST - SIN restricciones")
    print("  Solución:", solucion_test_sin)
    print("  Audiencia:", audiencia_test_sin)

    # Ejecutamos fuerza bruta CON restricción V20 y L20
    solucion_test_con, audiencia_test_con = fuerza_bruta(
        categorias_test,
        horarios_test,
        exigir_v20_l20=True
    )

    print("")
    print("TEST - CON restricciones V20/L20")
    print("  Solución:", solucion_test_con)
    print("  Audiencia:", audiencia_test_con)

# Ejecutamos el test
prueba_fuerza_bruta()
```

## Calcula la complejidad del algoritmo por fuerza bruta

### Coste de evaluar una combinación

Para cada combinación generada, el algoritmo realiza las siguientes operaciones:

1. **Construcción de la solución**  
   Se crea la lista `solucion = [...]` a partir del vector de índices.  
   Esto requiere recorrer \(n\) partidos para traducir cada índice al horario correspondiente → $O(n)$.

2. **Comprobación de la restricción V20/L20**  
   Se verifica si la solución contiene al menos un partido en los horarios `V20` y `L20`. En el peor caso se recorre toda la lista de horarios asignados → $O(n)$.

3. **Cálculo de la audiencia total (`calcular_audiencia_total`)**
   - Se recorre la lista de partidos para contar cúantos se disputan en cada franja horaria → $O(b)$
   - Se recorre nuevamente la lista de partidos para calcular su contribución a la audiencia total → $O(n)$

Por lo tanto, el coste de evaluar una combinación es aproximadamente:
$O(n + b)$

### Complejidad temporal total

El algoritmo de fuerza bruta explora todas las combinaciones posibles de horarios.

Dado que cada uno de los $n$ partidos puede asingarse a cualquiera de los $b$ horarios, el número total de combinaciones es $b^n$.

Como evaluar cada combinación cuesta $O(n + b)$, la complejidad temporal total es:

$T(n,b) = O(b^n*(n + b))$

Dado que en este problema el número de horarios b = 10 es constante, la complejidad suele expresarse como:
$O(b^n * n)$

### Conclusión

El algoritmo de fuerza bruta presenta **complejidad exponencial**, ya que el número de combinaciones crece como $b^n$.  

En el caso concreto del problema n = 10, b = 10, el espacio de búsqueda contiene más de **4.100 millones de combinaciones válidas**, lo que hace inviable ejecutar el algoritmo completo en la práctica. Por este motivo se utiliza un **caso de prueba reducido** para verificar su correcto funcionamiento.

## Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

```
def audiencia_base_partido(cat_local, cat_visit):
    """
    Devuelve audiencia base (millones) según categorías
    """
    return AUDIENCIA_BASE[(cat_local, cat_visit)]


def coef_correccion_por_m(m):
    """
    Devuelve el coeficiente multiplicativo por coincidencia en un slot.
    Si solo hay un partido, no penaliza. Si hay más de uno, penaliza acorde a PENALIZACION.

    Párametros:
    - m = número de partidos en un slot (int)

    Devuelve:
    - Coeficiente de corrección (float)
    """
    k = max(0, min(m - 1, 8))
    return 1.0 - PENALIZACION[k]


def contribucion_slot(p_h, m, sumaA):
    """
    Calcula la contribución total de un horario h al objetivo, agrupando:

    - p_h: ponderación del horario
    - m: nº de partidos en el horario
    - sumaA: suma de audiencias base de los partidos asignados a ese horario
    """
    if m <= 0:
        return 0.0
    return p_h * coef_correccion_por_m(m) * sumaA


def calcular_audiencia_total_por_slots(A, solucion_horarios):
    """
    Calcula el objetivo final a partir de una solución completa.

    A[i] = audiencia base del partido i.
    solucion_horarios[i] = horario asignado al partido i.
    """
    # Estructuras auxiliares por slot: cuántos partidos y suma de audiencias en cada slot.
    conteo = {h: 0 for h in HORARIOS}
    sumaA = {h: 0.0 for h in HORARIOS}

    # Acumulamos por horario lo que cae en cada slot.
    for i, h in enumerate(solucion_horarios):
        conteo[h] += 1
        sumaA[h] += A[i]

    # Sumamos la contribución de cada slot ya con su penalización por coincidencias.
    total = 0.0
    for h in HORARIOS:
        total += contribucion_slot(PONDERACION_HORARIO[h], conteo[h], sumaA[h])

    return total
```

```
# DEFINIMOS ALGORITMO VORAZ

def greedy_inicial(A):
    """
    Construye una solución rápida (no necesariamente óptima) para:
    - obtener una cota inferior inicial (best_aud)
    - cumplir las restricciones de que exista un partido en V20 y en L20

    - Fija los dos partidos de menos audiencia base a V20 y L20
    - El resto se asignan al slot que maximicen el incremento marginal (delta)

    """
    n = len(A)
    sol = [None] * n

    # Ordenamos los partidos con menor audiencia base
    idx_sorted = sorted(range(n), key=lambda i: A[i])

    # Forzamos que los 2 de menos audiencia sean en lunes y viernes para cumplir con la restricción
    if n >= 1:
        sol[idx_sorted[0]] = "V20"
    if n >= 2:
        sol[idx_sorted[1]] = "L20"

    conteo = {h: 0 for h in HORARIOS}
    sumaA = {h: 0.0 for h in HORARIOS}
    total = 0.0

    # Primero aplicamos los obligatorios actualizando incrementalmente el total.
    for i in range(n):
        if sol[i] is None:
            continue

        h = sol[i]
        p = PONDERACION_HORARIO[h]

        # before/after
        # contribución del slot antes de meter el partido i
        before = contribucion_slot(p, conteo[h], sumaA[h])
        conteo[h] += 1
        sumaA[h] += A[i]

        # Contribución del slot después de meter el partido i
        after = contribucion_slot(p, conteo[h], sumaA[h])

        # Delta sobre el total
        total += (after - before)

    # Para cada partido sin asignar, probamos todos los horarios y nos quedamos con el mejor delta
    for i in range(n):
        if sol[i] is not None:
            continue

        best_h = None
        best_delta = -1e18

        # Probamos todos los horarios y elegimos el que más aumenta el objetivo
        for h in HORARIOS:
            p = PONDERACION_HORARIO[h]

            before = contribucion_slot(p, conteo[h], sumaA[h])
            after = contribucion_slot(p, conteo[h] + 1, sumaA[h] + A[i])

            delta = after - before

            if delta > best_delta:
                best_delta = delta
                best_h = h

        # Confirmamos la mejor decisión encontrada y actualizo el estado del slot
        sol[i] = best_h
        p = PONDERACION_HORARIO[best_h]

        before = contribucion_slot(p, conteo[best_h], sumaA[best_h])
        conteo[best_h] += 1
        sumaA[best_h] += A[i]
        after = contribucion_slot(p, conteo[best_h], sumaA[best_h])

        total += (after - before)

    # Si no se cumpliesen las restricciones, invalidamos la solución
    if "V20" not in sol or "L20" not in sol:
        return None, -1e18

    return sol, total
```

```
# DEFINIMOS ALGORITMO DE RAMIFICACIÓN Y PODA

def branch_and_bound(partidos, team2cat):
    """
    Resuelve el problema por Branch & Bound.

    - Ramifica asignando un partido (en cierto orden) a un horario,
    - calcula una cota superior optimista para la rama,
    - poda si la cota superior no puede mejorar la mejor solución encontrada

    Devuelve (best_sol, best_aud):
    - best_sol[i] = horario asignado al partido i
    - best_aud = audiencia total máxima encontrada
    """
    n = len(partidos)

    # Construimos categorías y audiencias base A[i] para cada partido i
    categorias = categorias_de_partidos(partidos, team2cat)
    A = [audiencia_base_partido(*categorias[i]) for i in range(n)]

    # pmax es la mejor ponderación horaria posible
    pmax = max(PONDERACION_HORARIO.values())

    # Ordenamos los partidos por audiencia para intentar subir la cota inferior y podar más
    orden = sorted(range(n), key=lambda i: A[i], reverse=True)

    # llamamos al metodo greedy para comenzar con una solución decente
    best_sol, best_aud = greedy_inicial(A)
    if best_sol is None:
        best_sol, best_aud = None, -1e18

    conteo = {h: 0 for h in HORARIOS}
    sumaA = {h: 0.0 for h in HORARIOS}

    # sol guarda la asignación parcial/actual (backtracking)
    sol = [None] * n

    # remA[pos] = suma de audiencias base A de los partidos que quedan por asignar
    # ordenados por audiencia, desde pos hasta el final.
    # La usamos para construir la cota superior.
    remA = [0.0] * (n + 1)
    for pos in range(n - 1, -1, -1):
        remA[pos] = remA[pos + 1] + A[orden[pos]]

    def horas_posibles(pos, usado_v20, usado_l20):
        """
        Aplica una poda por restricciones.
        Si aún no he usado V20 y L20, necesito reservar capacidad para poder meterlos.

        - Si quedan menos de 2 partidos por asignar, es imposible cumplir.
        - Si queda 1 partido y no he usado V20 (o L20), fuerza ese horario.
        """
        restantes = n - pos

        # Si faltan ambos y quedan <2 => imposible
        if (not usado_v20) and (not usado_l20) and restantes < 2:
            return []

        # Si queda 1, forazamos lo que falte
        if restantes == 1:
          if (not usado_v20) and (not usado_l20):
            return []
          if (not usado_v20):
            return ["V20"]
          if (not usado_l20):
            return ["L20"]
          return HORARIOS

        return HORARIOS

    def delta_add(i, h):
        """
        Incremento marginal (delta) de asignar el partido i al horario h,
        usando el estado actual (conteo/sumaA) del slot.
        """
        p = PONDERACION_HORARIO[h]
        before = contribucion_slot(p, conteo[h], sumaA[h])
        after = contribucion_slot(p, conteo[h] + 1, sumaA[h] + A[i])
        return after - before


    def dfs(pos, total, usado_v20, usado_l20):
        """
        DFS sobre el árbol de decisiones.

        pos: nivel del árbol (cuántos partidos del 'orden' llevo asignados)
        total: valor objetivo acumulado hasta ahora (sumando deltas)
        usado_v20 / usado_l20: flags para asegurar la restricción de viernes/lunes
        """
        nonlocal best_aud, best_sol

        # Cota superior optimista:
        # asumimos que toda la audiencia base restante se puede colocar con ponderación pmax
        # (es optimista para no podar soluciones buenas).
        cota_sup = total + remA[pos] * pmax
        if cota_sup <= best_aud:
            return  # Si es mejor, podamos

        # Caso base: ya están todos los partidos asignados
        if pos == n:
            if usado_v20 and usado_l20 and total > best_aud:
                best_aud = total
                best_sol = sol.copy()
            return

        # Partido real a asignar en este nivel
        i = orden[pos]

        # Evaluamos que horarios nos quedan posibles
        hs = horas_posibles(pos, usado_v20, usado_l20)
        if not hs:
            return # Podamos en caso de no existir horas válidas

        # Ordenamos los horarios por delta decreciente para explorar primero las opciones más prometedoras
        # para tratar de aumentar la poda
        hs = sorted(hs, key=lambda h: delta_add(i, h), reverse=True)

        # Intentamos asignar el partido i a cada horario posible.
        for h in hs:
            d = delta_add(i, h)

            # Aplicamos decisión (forward step)
            sol[i] = h
            conteo[h] += 1
            sumaA[h] += A[i]

            dfs(
                pos + 1,
                total + d,
                usado_v20 or (h == "V20"),
                usado_l20 or (h == "L20"),
            )

            # Deshacemos decisión (backtrack)
            sumaA[h] -= A[i]
            conteo[h] -= 1
            sol[i] = None

    # Inicio de la búsqueda desde el estado vacío
    dfs(0, 0.0, False, False)
    return best_sol, best_aud
```

```
# EJEMPLO DE USO

sol, aud = branch_and_bound(PARTIDOS_EJEMPLO, TEAM2CAT)
print("Mejor solución (B&B):", sol)
print("Mejor audiencia (B&B):", aud)

# Comprobamos la audiencia obtenida
cats = categorias_de_partidos(PARTIDOS_EJEMPLO, TEAM2CAT)
A = [audiencia_base_partido(*cats[i]) for i in range(len(PARTIDOS_EJEMPLO))]
print("Audiencia verificada:", calcular_audiencia_total_por_slots(A, sol))
```

La **Fuerza Bruta** evalúa todas las asignaciones posibles de horarios, por lo que su coste crece exponencialmente con el número de partidos $n$.

En cambio, el método de **ramificación y poda** explora el mismo árbol de decisiones, pero en cada nodo calcula una cota superior del mejor valor que podría alcanzarse completando esa solución.

De este modo, si la cota superior de una rama es menor o igual que la mejor solución encontrada hasta el momento, esa rama se poda y no se continúa explorando, ya que no puede mejorar el resultado actual.

Así se evita evaluar una gran cantidad de combinaciones y, en la práctica, el algoritmo suele ser **mucho más eficiente** que en la exploración exhaustiva del árbol.

## Calcula la complejidad del algoritmo

En nuestro caso, el algoritmo de *branch and bound* (implementado con DFS) explora un árbol de asignaciones donde en cada nivel se asigna 1 partido a uno de los $h$ horarios disponibles.

Por lo tanto, en el peor caso (sin podas) el algoritmo podría visitar $h^n$ nodos (siendo n:cantidad de partidos); es decir, complejidad exponencial en $n$.

Además, en cada nodo se realizan varias operaciones:

- se calculo de una costa superior para la solución parcial,
- se evalua el incremento marginal (delta) de asignar el partido actual a cada uno de los $h$ horarios disponibles, y
- se ordenan esos horarios por su delta para explorar las opciones más prometedoras.

La ordenación de los $h$ horarios  cuesta $O(h\log(h))$, por lo que el coste total en el peor caso queda acotado por

$$
O(h^n \cdot h\log h)=O(h^{n+1}\log h),
$$

A esto se le suma el coste inicial del algoritmo greedy utilizado par obtener una primera solución (cota inferior), con un coste aproximado de: $O(n\log n + n\,h)$


En la práctica, al podar las ramas cuyo cota superior no mejora la mejor solución encontrada, el número real de nodos explorados puede caer muchísimo respecto a $h^n$, aunque el **peor caso siga siendo exponencial**.


## Datos de entrada aleatorios

Según el problema (y tenga sentido), diseña un juego de datos de entrada aleatorios

```
import random

def partidos_jornada(equipos):
    """
    Genera una jornada aleatoria emparejando los equipos en partidos.

    Párametros:
    - Equipos: lista de equipos a emparejas (debe tener tamaño par).

    Devuelve:
    - Lista de tuplas (partidos = (equipo1,equipo2))
    """

    if len(equipos) % 2 != 0:
        raise ValueError("Debe haber un número par de equipos")

    equipos_sin_asignar = equipos[:]
    random.shuffle(equipos_sin_asignar)
    partidos = []

    while equipos_sin_asignar:
        equipo1 = equipos_sin_asignar.pop()
        equipo2 = equipos_sin_asignar.pop()
        partidos.append([equipo1, equipo2])
    return partidos
```

```
equipos = ['Real Madrid', 'Barcelona', 'R.Sociedad','Celta', 'Valencia', 'Athletic',
           'Villarreal', 'Alavés', 'Levante', 'Sevilla', 'Espanyol', 'Betis', 'Atlético',
           'Getafe','Mallorca', 'Eibar', 'Leganés', 'Osasuna', 'Granada', 'Valladolid']

print(partidos_jornada(equipos))
```
